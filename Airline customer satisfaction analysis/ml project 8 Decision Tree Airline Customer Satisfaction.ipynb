```python
from sklearn.preprocessing import LabelEncoder
```

```python
model=LabelEncoder()
```

```python
data[['Gender','Customer_Type','Type_of_Travel','Class']]=data[['Gender','Customer_Type','Type_of_Travel','Class']].apply(model.fit_transform)
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
x=data.drop('satisfaction',axis=1)
y=data['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=22)
```

```python
data.hist();
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x_train)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='gini',splitter='random',max_depth=3,min_samples_split=2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
print(classification_report(y_test,model.predict(x_test)))
```

```python
sns.heatmap(confusion_matrix(y_test,model.predict(x_test)),annot=True);
```
